```python
import csv 
import pandas as pd
```

RUN THE BELOW CODE TO CREATE A FILE FOR THE SETING ARRANGEMENT BLUEPRINT

```python
no_of_seats = int(input("Enter the number of seats in class: "))
no_of_class = int(input("Enter the number of classes in school: "))
print(f"Total capacity of Student: {no_of_seats*no_of_class}")

test = [''for i in range(no_of_seats)]
test = [test for i in range(no_of_class)]
seats = [f"Seat-{i}" for i in range (1,no_of_seats+1)]

with open ("test.csv","w",newline='') as file:
    writer = csv.writer(file)
    writer.writerow(seats)
    writer.writerows(test)
```

RUN THE BELOW CODE TO ADD STUDENT IN CLASS AND CLASS SPECIFIC SEAT

```python
df = pd.read_csv("test.csv")
total_student = int(input("Enter the number of student u have: "))

if total_student <= no_of_class * no_of_seats:

    seting_plan = []

    if total_student % no_of_seats == 0:

        seting_plan.append((int(total_student/no_of_seats), 0))

    else:

        reminder = total_student % no_of_seats
        coefficent = int((total_student-reminder)/no_of_seats)

        seting_plan.append((coefficent, reminder))


    for row, col in seting_plan:

        seat = 1

        # fill full classes
        for i in range(row):

            for j in range(no_of_seats):

                df.iloc[i, j] = seat
                

                seat += 1


        # fill remainder students (only if remainder exists)
        if col != 0:

            for j in range(col):

                df.iloc[row, j] = seat
                seat += 1
    df = df.fillna(0).astype(int)      
    df.to_csv("test.csv", index=False)
    print("Data added")

else:

    print("Not enough seats")
    
```

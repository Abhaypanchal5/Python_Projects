```python
import pandas as pd
import csv
```

Run the below code to create an Indexed csv file to store and access the Question and crossponding Choices

```python
index = []

index.append("Question")

for i in range(1,5):
    index.append(f"Option-{i}")
    
index.append("Answer")

with open ("Question.csv","w",newline="") as file:
    writer = csv.writer(file)
    writer.writerow(index)
```

 RUN THE BELOW CODE TO INSERT THE QUESTIONS AND ITS ANSWER WITH FOLLOWING CHOICES

```python
n = int(input('Enter the number of question u want to enter: '))
for i in range(1,n+1):
    Q = []
    question = input(f"Enter your question {i}: ")
    Q.append(question)
    for i in range (1,5):
        option = input(f"Enter your {i} option: ")
        Q.append(option)
    answer = input("Enter the answer of the question: ")
    Q.append(answer)
    with open ("Question.csv","a",newline="") as file:
        writer = csv.writer(file)
        writer.writerow(Q)
```

RUN THE BELOW CODE TO RETRIVE THE QUESTION AND GET THE ANSWER FROM THE USER AND STORE IT IN A LIST

```python
data = pd.read_csv("Question.csv")
Student_answer = []
no_of_question = 3    # can be found running len funtion on the data
for j in range(1, no_of_question + 1): 
    print(f"Question {j}) {data.iloc[j-1, 0]}")
    for i in range(1, 5):
        print(f"option-{i}) {data.iloc[j-1, i]}")
    answer = input("Enter your Answer: ")   
    Student_answer.append(answer)
```

USE THE BELOW CODE TO CALCULATE IS THE ANSWER IS CORECT OR NOT

```python
result = []
for j in range(2):
    student_list = []
    for i in range(3):
        if data.loc[i,"Answer"] == Student_Ans_data.iloc[j,i]:
            student_list.append(1)
        elif data.loc[i,"Answer"] != Student_Ans_data.iloc[j,i]:
            student_list.append(-0.25)
        else:
            student_list.append("")
    result.append(student_list)
result
```

USE BELOW CODE TO ADD DATA TO THE FILE

```python
with open("Student_marks.csv", "a", newline="") as file:
    writer = csv.writer(file)
    
    writer.writerows(result)
```

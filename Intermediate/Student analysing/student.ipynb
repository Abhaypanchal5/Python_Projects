```python
import pandas as pd
import csv
```

```python
def Add_data(data):
    if isinstance(data, str):
        data = [[data]]
    with open ("Result.csv","a",newline='') as file:
        writer = csv.writer(file)
        writer.writerows(data)
    
```

```python
data = pd.read_csv("student.csv")
```

RUN THE BELOW CODE TO ADD NEW DATA COLOUM 

```python
data["Total"] = data[["Math","Science","English"]].sum(axis=1)
data["Avg"] = data["Total"]/3 #no of subject
```

CODE TO FIND THE TOPPER

```python
data = data[["Name","Total"]].sort_values("Total",ascending = False)
print(f"{data.iloc[0,0]} is the topper with {data.iloc[0,1]}")
```

```python
Add_data("Topper: Sneha")
```

```python
def check_no_of_student_fail_in_sub(Subject_name,passing_marks=50):
    failed = data[Subject_name]>=passing_marks
    return print((~failed).sum())
    
```

RUN THE BELOW TWO CODE TO GET THE FAIL STUDENT

```python
failed_student = []

# Use range(len(data)) to stay within bounds
for j in range(len(data)):
    # Check columns starting from index 1 (assuming index 0 is the Name/ID)
    for i in range(1, len(data.columns)):
        if data.iloc[j, i] < 50:
            failed_student.append(data.iloc[j, 0])
            break # Stop checking this student once they've failed one subject
```

```python
# This finds any student who has at least one score below 50 
# and returns the values from the first column (index 0)
failed_student = data[data.iloc[:, 1:].lt(50).any(axis=1)].iloc[:, 0].tolist()
```

```python
Add_data(f"Failed student: {', '.join(failed_student)}")
```

```python
total_class_marks = data["Total"].sum()

class_avg = total_class_marks/len(data)

class_avg
```

```python
Add_data(f'Class average: {class_avg}')
```

```python
def subject_topper(subject):
    for i in range (len(data)):
        if data.loc[i,subject] == data[subject].max():
            return data.loc[i,"Name"]    
```

```python
Add_data(f"Math topper: {subject_topper("Math")}")
Add_data(f"Science topper: {subject_topper("Science")}")
Add_data(f"English topper: {subject_topper("English")}")
```

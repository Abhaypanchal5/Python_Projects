```python
# This cell generate email based on user input(First name,last name) and add it to the file generated by privious cell print the conformation massage once email is generated

import csv
import random

Fname = (input("Enter Your First Name: ")).strip()
Lname = (input("Enter Your Last Name: ")).strip()
ran = ''.join(random.choices("1234567890",k=4))
Email = Fname.lower() + Lname.lower() + ran  + '@gmail.com'
with open ("Generated_email.csv",'a',newline="") as file:
    writer = csv.writer(file)
    writer.writerow([Fname,Lname,Email])
    
print("The email is generated sucessfully")
print(f"The system generated email: {Email}")
    
```

```python
# This Cell show the system generated email to user when required and give an error if email is not present

import csv
fname = (input("Enter the First name")).strip
lname = (input("Enter the last name: ")).strip
with open ("Generated_email.csv") as f:
    reader = csv.DictReader(f)
    for row in reader:
        if row["Fname"] == fname and row["Lname"] == lname:
            print("the Registered email in the system is: ",row["email"])
            exit()
    else:
         print('Email not found!!!')
```

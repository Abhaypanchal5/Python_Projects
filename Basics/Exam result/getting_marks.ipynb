```python
df = pd.read_csv("Answers.csv")          #read the answers file 
af = pd.read_csv("student_answer.csv")   #read the students answers
```

```python
# transpose the data for better reading

af = af.T
df = df.T
```

```python
total = []
for j in range (2):   # here range is the number of student(u can change it according to data)(l)
    student = []
    for i in range (1,11):  # here range is the no of question (in my case its 10 question) chamge according to the question number length(n)
        if df.loc[str(i),0] == af.loc[str(i),j]:  
            student.append(1)
        elif df.loc[str(i),0] != af.loc[str(i),j]:
            student.append(-0.25)
        else:
            student.append(0)
    total.append(student)       # appending the student mark in total list indivisually
```

```python
result = [] #in list indexing is done according to student example student 1 = index 0

for x in range (2):   # here n is the no of student 
    result.append(sum(total[x]))  
    
result #printing the result of all student
```
